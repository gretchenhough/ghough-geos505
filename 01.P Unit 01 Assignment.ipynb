```python
def IsLeapYear(year): #this is creating a function
    
    if (year % 4 == 0): #if statement that will be true if the input into the function has a factor of 4
        if (year % 500 == 0): #if statement that will be true if input has a factor of 500
            return True #this marks the value as true
        elif (year % 100 == 0):#if 500 isn't a factor but 100 is...
            return False #the value is marked as false
        else: #if only 4 is a factor it will also mark value as true
            return True
    else: #anything not factorable by 4 is marked false
        return False

## the function IsLeapYear is supposed to identify which years are leap years so Feburary can be assigned the correct number of days. HOWEVER! The rule is must be divisible by 400 NOT 500, so this function will not return the correct leap years (thank you, Google!)

def ReturnFileList(file_base, file_ext, start_yr, end_yr): #defining a new function that will create a list of files, you need to input the file prefix, file extension, what year you want to start, and what year you want to end when you use this function.
    
    DiM = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31] #DiM="days in month" this is an array of 12 months with the appropriate number of days assigned assuming no leap year

    year = start_yr #setting the counter to your start value

    files = [] #creating a blank array to write stuff to called "files"
    
    for i in range(end_yr - start_yr + 1): # running a loop that will take a step for each year from start to end +1 to capture end year
        for j in range(12): #loop to check for leap years, range 12 because 12 months, so j is counting months looping through all 12 inside one step of the outer i loop
            if (j+1==2) and IsLeapYear(year): #when j=2 (set as j+1 because counter starts at 0), check if the leap year function works
                days_in_month = 29 #if it is Feb of a leap year, Feb will be assinged 29 days
            else: #if it's not a leap year, or not Feb, it will be assigned the value in the jth position of the array DiM
                days_in_month = DiM[j]
            
            for k in range(days_in_month): #running another internal loop counting each day in the month
                
                file_name = f'{file_base}-{year:4d}-{(j+1):02d}-{(k+1):02d}{file_ext}' #writing the file name for each day in the month
                files.append(file_name) #adding the file name to the array "files"
                
        year = year + 1 #moving on to the next year for the next step of the outer loop!
    
    return files #when the function is done running the loops, it will return and print the "files" array
```

```python
StartYr = 1998 #Don't need to define these outside of the function. Can just put these as years in the function itself.
EndYr = 2000

file_list = ReturnFileList('precip','.nc',StartYr,EndYr) #The function ReturnFileList already creates a list of files under "files." This is renaming it and not necessary.

print(str(len(file_list))+' files') #This will report back the number of files generated followed by the word "files."
print(*file_list, sep='\n') #This is printing a list of the file names. The * makes it so it treats each item in the list seperatly for the print command. The sep='\n' puts each file name on a new line.
```
